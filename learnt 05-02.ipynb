JSON

To convert a JSON string to Python form, use json.loads. json.dumps on the other hand converts a Python object back to JSON. Because the JSON structure is unpredictable, how you convert a JSON object or list of objects to a DataFrame or some other data structure for analysis will be up to you. Conveniently, we are able to pass a list of JSON objects to the DataFrame constructor and select subsets of fields.

```python
import requests

# Endpoint for retrieving Yahoo Weather for Toronto, ON. 
apiEndpoint =  'http://ws1.metcheck.com/ENGINE/v9_0/json.asp?lat=43.7&lon=-79.4&lid=4868&Fc=No'

'''apiEndpoint Definition: 
This line defines a variable named apiEndpoint which contains the URL of an API endpoint. The URL points to a specific location for weather data. The endpoint is provided by Metcheck and is formatted to return JSON data for a specific latitude and longitude (in this case, latitude 43.7 and longitude -79.4).'''


response_json = pd.read_json(apiEndpoint, lines=False)
'''lines : bool, default False
    Read the file as a json object per line.'''

'''response_json Assignment: 
This line sends a request to the API endpoint defined in apiEndpoint using the pd.read_json() function from the pandas library. This function reads JSON data from the specified URL and converts it into a pandas DataFrame. '''

response_json
```

```python
pd.DataFrame(response_json["metcheckData"]["forecastLocation"])

'''response_json["metcheckData"]["forecastLocation"]: 
This part of the code accesses the nested structure of the JSON response. It retrieves the data under the keys "metcheckData" and "forecastLocation", which likely contain information specific to the forecast location.'''
```
